```python
stok_data_filtered = stok_data.loc[:, ["Stok Kodu", "L.Fiy. 1", "L.Fiy. 2", "L.Fiy. 3", "L.Fiy. 4", "L.Fiy. 5"]]
stok_data_filtered.head()
```

```python
stok_data_filtered = stok_data_filtered.rename(columns={"Stok Kodu": "BARKOD"})
stok_data_filtered.head()
```

```python
stok_data_filtered = stok_data_filtered[stok_data_filtered['BARKOD'].isin(ticimax_data['BARKOD'])]
stok_data_filtered.head()
```

```python
stok_data_lst = stok_data_filtered.to_dict('records')
print(stok_data_lst)
len(stok_data_lst)
```

```python
def process_stock_data(stok_data_lst):
    result = {}
    for stok_data in stok_data_lst:
        new_stok = dict()
        new_stok["BARKOD"] = stok_data["BARKOD"]
        for i in range(1, 6):
            field_name = f"L.Fiy. {i}"
            if stok_data[field_name] != "0,00 ₺":
                new_stok["Price"] = stok_data[field_name]
        if new_stok.get("Price") is None:
            new_stok["Price"] = ""
        result[new_stok["BARKOD"]] = new_stok
    return result
stok_data_dict = process_stock_data(stok_data_lst)
```

```python
import csv

# specify the file path and name
file_name = 'output.csv'

# specify the dictionary to write to CSV
my_dict = ticimax_data.to_dict("records")

# open the file in write mode
with open(file_name, mode='w', newline='') as file:
    writer = csv.DictWriter(file, fieldnames=my_dict[0].keys())
    writer.writeheader()
    writer.writerows(my_dict)
```
